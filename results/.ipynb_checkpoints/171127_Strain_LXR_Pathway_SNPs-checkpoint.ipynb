#### Get list of genes expressed in the strains
____  
To do this we use annotate repeats. HOMER then annotates the genes that are expressed, helping us create a data file that we can use moving forward to filter the genes and create a bed file

```python
# Run annotate repeats for exons - rpkm normalized for conversion to tpm
# For now this rpkm file is only used to convert the gene names from GO into refseq NM mrna annotations
if not os.path.isfile(outdir + '/rnaseq_matrixGenes_rpkm.txt'):
    out, err = hbUtils.run_annotateRepeats(samplepaths, countType = 'genes', norm='rpkm')
    f = open(outdir + '/rnaseq_matrixGenes_rpkm.txt', 'w')
    f.write(out.decode('utf-8'))
    f.close()
```

```python
# open annotate repeats file and process
rpkm = pd.read_csv(outdir + '/rnaseq_matrixGenes_rpkm.txt', sep='\t')

# Create column for gene
rpkm['Gene'] = pd.Series([x.split('|')[0] for x in rpkm['Annotation/Divergence']])

# clean up the columns a bit
pattern = r"\s+\(.+\)$"
rpkm.columns = pd.Series(rpkm.columns)\
        .str.replace(pattern, '')\
        .str.split('/')\
        .str[-1]
```

#### Pull gene list we are interested in from QuickGO database  
____  
for now that is:  
* GO:0033344 cholesterol efflux
* GO:0006633 fatty acid biosynthetic process  

Consider in the future:  
* GO:0008610 lipid biosynthetic process

```python
# import using pull_go_terms function, accepts GO term as well as taxon id as a string
# 10090 is mus musculus
chol_eff = pd.DataFrame(hbUtils.pull_go_terms('GO:0033344', taxon_id='10090'))[1]\
        .drop_duplicates().dropna()
fa_biosyn = pd.DataFrame(hbUtils.pull_go_terms('GO:0006633', taxon_id='10090'))[1]\
        .drop_duplicates().dropna()

# combined list of genes in these pathway
gene_list = pd.concat([chol_eff, fa_biosyn])
```

#### Use annotate repeats file to convert gene name to the NM identifier

```python
# Generate BED file using the annotateRepeats file
# create pattern to match all the genes with pd.DataFrame.str.match
gene_regex = r'\||'.join(gene_list)
rpkm_subset = rpkm[rpkm['Divergence'].str.match(gene_regex, flags=re.IGNORECASE)]

# write gene id file
rpkm_subset['RepeatID']\
        .drop_duplicates()\
        .dropna()\
        .to_csv(outdir + 'chol_eff_fa_biosyn_gene_id.txt',
                header=False, index=False, sep='\t')
```

#### Get mutation calls to feed into JCVI provean
_____
This is a pretty lazy solution, you have to copy the list of variables then paste it into JCVI's web based service at:  
http://provean.jcvi.org/genome_submit_2.php?species=mouse  
however it is a fairly powerful tool and allows good annotation so we'll go with that for now.  
**files will need to be scp'd back onto epiglass into results folder as**
* BALBcJ_Provean_Results.tsv 
* AJ_Provean_Results.tsv  

```python
# define function for processing through annotate_mutations mut files
def format_mut(tst, col):
    
    # split into list of all variants
    variants = tst.iloc[col].split(',')
    
    # for each variant print out a string we can feed to provean
    for i in variants:
        print(tst.chr.replace('chr', '') + ',' + i\
            .replace(':', ',')\
            .replace('->', ','))
```

```python
exon_muts[exon_muts['AJ -1'].notnull()].apply(format_mut, col=-2, axis=1)
```

```python
exon_muts[exon_muts['BALBCJ - 1'].notnull()].apply(format_mut, col=-1, axis=1)    
```

```python
# read in files
aj_prov = pd.read_csv(outdir + 'AJ_Provean_Results.tsv', sep='\t')
balb_prov = pd.read_csv(outdir + 'BALBcJ_Provean_Results.tsv', sep='\t')
```

### Interpretation:

There are some predicted deleterious mutations within the strains in the lipid and cholesterol processing pathways. Interestingly the only deleterious mutations that were predicted by both SIFT and Provean were in balb, and balb were the mice that DID NOT get fatty liver disease while they did develop obesity. These occured in Aloxe3 and Ptplad2.
* **Ptplad2 (HCAD4)** 3-Hydroxyacyl-CoA Dehydratase 4 / Protein-Tyrosine Phosphatase-Like A Domain-Containing Protein 2
    * Involved in fatty acid elongation and metabolism, extends palmitoyl-CoA and stearoyl-CoA to very long chain acyl CoAs. Third component of microsomal "elongase", however many homologous genes in mice http://www.genome.jp/dbget-bin/www_bget?K10703+K18880+4.2.1.134+R10827
    * Low expression in adult liver (Encode transcriptome ncbi). Expressed in macrophages (immgen rna microarray)
    * Suthar et al 2013 found that Ptplad2 was regulated by Type 1 IFN and RIG-1 like receptor (RLR - recognizes viral RNA) in liver tissues. http://journals.plos.org/plospathogens/article?id=10.1371/journal.ppat.1003168
* **Aloxe3**
    * Krieg et al 2013 found that ALOXE3 null mice were lethal due to skin barrier dysfunction.
http://www.sciencedirect.com/science/article/pii/S0022202X15359480?via%3Dihub
    * almost entirely expressed in skin and in alveolar macrophages.  
    * most known functions in forming waxy vlcfa's for stratum corneum
  
Other predicted deleterious mutations...
* **Lpgat1** (Balbc and AJ - insertion, AJ has two insertions, all non-frameshift)
    * reacylation of lysophosphatidylglycerol to phosphatidylglycerol (membrane phospholipid)
    * some expression in mouse macrophages. expression in many tissues.
    * involved in triacylglycerol synthesis monoacylglycerol pathway, found to have mono-acylglycerol acyl-transferase activity (MGAT) to go from Monoacyl glycerol --> diacylglycerol (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3035505/)
* **Egf** (Balbc and AJ same mutation):
    * balbc and aj are not reported to have altered egf signaling
* **Apoc3** (Balbc and AJ same mutations):
    * VLDL protein, inhibits LPL and hepatic lipase. delays catabolism of G rich particles
    * overexpression appears to induce hypertriglyceridemia and NAFLD
* **Apoa2** (AJ)
    * encodes component of HDL. Mice lacking this protein have low HDL, increased TG rich lipoprotein clearance, and insulin hypersensitivity.
    * Could be defect in reverse cholesterol transport pathway?
* **Apob** (Balb)
    * Key protein of LDL and chylomicrons.
    * however knockdown of this protein still shows strong changes in LDL and TG content https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3090230/ so it could still be working
